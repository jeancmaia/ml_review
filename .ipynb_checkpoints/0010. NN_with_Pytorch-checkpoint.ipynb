# DL for classification and regression



```python
# https://www.learnpytorch.io/02_pytorch_classification/

from sklearn.datasets import make_circles

n_samples = 3000

X, y = make_circles(n_samples, noise=0.05, random_state=42) 
```

```python
# Visualize with a plot
import matplotlib.pyplot as plt
plt.scatter(x=X[:, 0], 
            y=X[:, 1], 
            c=y, 
            cmap=plt.cm.RdYlBu);
```

```python
# Split data into train and test sets
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, 
                                                    test_size=0.3, 
                                                    random_state=42) 

len(X_train), len(X_test), len(y_train), len(y_test)
```

```python
import torch
from torch import nn

batch_size = 10

X_train__tensor = torch.from_numpy(X_train).to(torch.float32)
y_train__tensor = torch.from_numpy(y_train).to(torch.float32)
X_test__tensor = torch.from_numpy(X_test).to(torch.float32)
y_test__tensor = torch.from_numpy(y_test).to(torch.float32)


train_tensor_dataset = torch.utils.data.TensorDataset(X_train__tensor, y_train__tensor)
test_tensor_dataset = torch.utils.data.TensorDataset(X_test__tensor, y_test__tensor)


trainloader = torch.utils.data.DataLoader(train_tensor_dataset, batch_size=batch_size,
                                          shuffle=True, num_workers=2)

testloader = torch.utils.data.DataLoader(test_tensor_dataset, batch_size=batch_size,
                                         shuffle=False, num_workers=2)
```

```python
device = "cuda" if torch.cuda.is_available() else "cpu"
device
```

```python
class CircleModel(nn.Module):

    def __init__(self):
        super().__init__()
        
        self._model = nn.Sequential(
            nn.Linear(in_features=2, out_features=10),
            nn.ReLU(),
            nn.Linear(in_features=10, out_features=1),
            nn.Sigmoid()
        )
    
    def forward(self, x):
        return self._model(x) 

model_circle = CircleModel()
```

```python
import torch.optim as optim

criterion = nn.BCELoss()
optimizer = optim.Adam(model_circle.parameters(), lr=0.001)
```

```python
epochs = 100

loss_values = []
test_roc_auc_stats = []
epochs_list = []

for epoch in range(epochs): 

    running_loss = 0.0
    for i, data in enumerate(trainloader, 0):

        inputs, labels = data
        # zero the parameter gradients
        optimizer.zero_grad()

        # forward + backward + optimize
        outputs = model_circle(inputs)
        loss = criterion(outputs, labels.unsqueeze(1))
        loss.backward()
        optimizer.step()

        # print statistics
        running_loss += loss.item()

        if i % 50 == 49:    # print every 100 mini-batches
            print(f'[{epoch + 1}, {i + 1:5d}] loss: {running_loss / 50:.3f}')
            loss_values.append(running_loss/ 50)
            test_roc_auc_stats.append(roc_auc_score(
                y_test,
                model_circle(X_test__tensor).tolist() 
            ))
            epochs_list.append(epoch)
            running_loss = 0.0

print('Finished Training')
```

```python
# plot lines
plt.plot(epochs_list, loss_values, label = "Binary CrossEntropy")
plt.plot(epochs_list, test_roc_auc_stats, label = "ROC_AUC Test")
plt.xlabel("Epochs")
plt.legend()
plt.show()
```

```python
from sklearn.metrics import roc_auc_score

roc_auc_score(
    y_test,
    model_circle(X_test__tensor).tolist() 
)
```

```python
from sklearn.metrics import precision_recall_curve, PrecisionRecallDisplay


#precision, recall, thresholds = precision_recall_curve(y_test, model_0(X_test__tensor).tolist() )
PrecisionRecallDisplay.from_predictions(y_test, model_circle(X_test__tensor).tolist() )
```

```python
from sklearn.calibration import calibration_curve, CalibrationDisplay

prob_true, prob_pred = calibration_curve(y_test, model_circle(X_test__tensor).tolist(), n_bins=5)

disp = CalibrationDisplay(prob_true, prob_pred, y_test)
disp.plot()
```

# Transfomers

It empowers the NN by coping with long passages of differing lengths and connections between word representations, therefore providing a method to fit the dependence throughout observations/tokens. In NLP, it is possible to observe encoder, decoder and encoder-decoder architectures. They are used to transform data into embeddings representation, generative components and sequence to sequence tasks. To cite some interesting implementations:
