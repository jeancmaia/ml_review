#### Conformal Prediction   


https://leanpub.com/conformal-prediction    
https://arxiv.org/pdf/2107.07511.pdf   

Conformal prediction is one of the cutting-edge tools for measuring models prediction uncertainty. This apparatus is key when we aim to craft robust machine learning systems and can be applied in both regression and classification model, in an agnostic fashion. It is agnostic and distribution-free.


To introduce the topic, imagine that we develop a model that predicts the tuple (cat, dog) for images. Since it is a classification problem, we could simply leverage the `predict_proba` for measuring the uncertainty, right? Sometimes not. Despite the name, that "proba" may not mean a consistent probability of occurrence, especially for non-calibrated models. 

`Conformal Prediction` transforms an uncertainty score into a rigorous score, with guaranteed coverage. It changes point predictions into prediction regions. For regression models, it returns a confidence interval, whereas a set with classes for classification.


Let's take a peek at a multi-classification model.

```python
import os
import wget
import zipfile
import pandas as pd
import numpy as np


from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from lightgbm import LGBMClassifier

from os.path import exists

# Download if not available
bean_data_file = "./DryBeanDataset/Dry_Bean_Dataset.xlsx"
base = "https://archive.ics.uci.edu/ml/machine-learning-databases/"
dataset_number = "00602"
if not exists(bean_data_file):
    filename = "DryBeanDataset.zip"
    url = base + dataset_number + "/" + filename
    wget.download(url)
    with zipfile.ZipFile(filename, 'r') as zip_ref:
        zip_ref.extractall('./')
    os.remove(filename)
```

```python
lgbm_model = LGBMClassifier()

# Read in the data from Excel file
bean_data_file = "./DryBeanDataset/Dry_Bean_Dataset.xlsx"
beans = pd.read_excel(bean_data_file)

# Labels are characters but should be integers for sklearn
le = LabelEncoder()
beans["Class"] = le.fit_transform(beans["Class"])

# Split data into classification target and features
y = beans["Class"]
X = beans.drop("Class", axis = 1)

# Split of training data
X_train, X_rest1, y_train, y_rest1 = train_test_split(
X, y, train_size=10000, random_state=2
)

# From the remaining data, split of test data
X_test, X_rest2, y_test, y_rest2 = train_test_split(
X_rest1, y_rest1, train_size=1000, random_state=42
)

# Split remaining into calibration and "new" data
X_calib, X_new, y_calib, y_new = train_test_split(
X_rest2, y_rest2, train_size=1000, random_state=42
)

# Fit the model
lgbm_model.fit(X_train, y_train)
```

```python
from yellowbrick.classifier import ROCAUC
from yellowbrick.classifier import PrecisionRecallCurve


visualizer = ROCAUC(lgbm_model, classes=le.classes_.tolist())

visualizer.fit(X_train, y_train)       
visualizer.score(X_test, y_test)      
visualizer.show()                 
```

```python
# Create the visualizer, fit, score, and show it
viz = PrecisionRecallCurve(
    lgbm_model,
    per_class=True,
    classes=le.classes_.tolist(),
    cmap="Set1"
)
viz.fit(X_train, y_train)
viz.score(X_test, y_test)
viz.show()
```

```python
predict_proba_test = lgbm_model.predict_proba(X_test)
```

```python
### Calibration metric.
from sklearn.metrics import brier_score_loss

for index in range(predict_proba_test.shape[1]):
    print('brier score for class ', 
          index,
          '(', le.classes_.tolist()[index], ')',
          ': ',
          brier_score_loss(y_test == index, predict_proba_test[:,index]))
```

```python
## Analyzing Calibration Curve of two lowest brier score classes
```

```python
from sklearn.calibration import CalibrationDisplay

disp = CalibrationDisplay.from_predictions(y_test == 3, predict_proba_test[:,3], n_bins = 10, name=le.classes_.tolist()[3])
```

```python
disp = CalibrationDisplay.from_predictions(y_test == 6, predict_proba_test[:,6], n_bins = 10, name=le.classes_.tolist()[6])
```

We decide to predict with 95% confidence. Despite the lack of calibration, filtering out the top 0.95 of `predict_proba` would be the simplest solution. Let's analyze the coverage of .95, a.k.a the ground truth in this region.

```python
# Get the "probabilities" from the model
probas_calib = lgbm_model.predict_proba(X_calib)

# Get for each instance the highest probability
high_prob_predictions = np.amax(probas_calib, axis=1)

# Select the predictions where probability over 95%
high_p_beans = np.where(high_prob_predictions >= 0.85)

# Let's count how often we hit the right label
its_a_match = (lgbm_model.predict(X_calib) == y_calib)
coverage = np.mean(its_a_match.values[high_p_beans])
print(round(coverage, 3))
```

The threshold 0.95 doesn't entail in 95% probability. How can we get a 95% coverage guarantee? Let's leverage probability to construct a measure of uncertainty.   

$$s = 1 - f(x_i)$$

where $f()$ is the predict_proba of our model. This s-score is called non-conformity score. It points out how unusual a suggested outcome y given the model output. Conformal prediction uses the s-score to calculate the threshold.

```python
# Get the probability predictions
y_prediction_proba = lgbm_model.predict_proba(X_calib)

## Get conformity
conformity_scores = np.take_along_axis(
                    1 - y_prediction_proba, y_calib.values.reshape(-1, 1), axis=1
                )


# Size of calibration data
n = len(conformity_scores)

confidence_interval = 0.95

### Get quantiles
quantiles = np.quantile(conformity_scores, ((n + 1) * confidence_interval) / n, method="linear")



y_prediction__new = lgbm_model.predict_proba(X_new)
prediction_sets = np.greater_equal(
                    y_prediction__new, (1 - quantiles)
            )
```

```python
y_set = np.squeeze(prediction_sets)
pd.Series(y_set.sum(axis=1)).value_counts(normalize=True)
```

```python
from mapie.metrics import classification_coverage_score
from mapie.metrics import classification_mean_width_score

cov = classification_coverage_score(y_new, y_set)
setsize = classification_mean_width_score(y_set)
print('Coverage: {:.2%}'.format(cov))
print("Avg. set size: {:.2f}".format(setsize))
```

```python
from mapie.classification import MapieClassifier
```

```python
mapie_score = MapieClassifier(lgbm_model, cv='prefit', method='score')
mapie_score.fit(X_calib, y_calib)
y_pred, y_set = mapie_score.predict(X_new, alpha=0.05)
y_set = np.squeeze(y_set)
```

```python
cov = classification_coverage_score(y_new, y_set)
setsize = classification_mean_width_score(y_set)
print('Coverage: {:.2%}'.format(cov))
print("Avg. set size: {:.2f}".format(setsize))
```

### Regression Confidence interval

The simplest case is when we conformalize the mean regression.

```python
# Read the rent index data
rent = pd.read_csv("http://www.bamlss.org/misc/rent99.raw", sep=' ')
y = rent["rentsqm"]
X = rent.drop(["rent", "rentsqm", "cheating"], axis = 1)
# Split of training data
X_train, X_rest1, y_train, y_rest1 = train_test_split(
X, y, test_size=2000, random_state=2
)
# Split of test data
X_test, X_rest2, y_test, y_rest2 = train_test_split(
X_rest1, y_rest1, test_size=1500, random_state=4
)
# Split rest into calibration and new data
X_calib, X_new, y_calib, y_new = train_test_split(
X_rest2, y_rest2, test_size=500, random_state=42
)
"Data sizes: train: %i, test: %i, calibration: %i, new: %i" % (
len(X_train), len(X_test), len(X_calib), len(X_new)
)
```

```python
from lightgbm import LGBMRegressor

lgbm_model = LGBMRegressor()

lgbm_model.fit(X_train, y_train)
```

```python
from sklearn.metrics import mean_absolute_error
# Make predictions using the fitted model
y_pred = lgbm_model.predict(X_test)
# Calculate the mean absolute error
mae = mean_absolute_error(y_test, y_pred)
print(round(mae, 2))
```

```python
from mapie.regression import MapieRegressor
mapie_reg = MapieRegressor(estimator=lgbm_model, cv="prefit")
mapie_reg.fit(X_calib, y_calib)
```

```python
y_pred, y_pis = mapie_reg.predict(X_new, alpha=0.05)
```

```python
interval = y_pis[0].flatten()
print("Predicted rent: {:.2f}".format(y_pred[0]))
print("95% interval: [{:.2f};{:.2f}]".format(interval[0], interval[1]))
```

```python
# Plots the quantile range of predicted values for given feature.
def plot_quantile_by_feature(X, y, pred, qrs, feature_name):
    """
    - X (pandas DataFrame): Dataframe with features for the data.
    - y (pandas Series): Pandas series with the target variable.
    - pred (array-like): An array with predictions.
    - qrs (array-like): Quantile range for plotting.
    - feature_name (string): The name of the feature being plotted.
    """
    xj= X[feature_name]
    # Sort the feature values
    order = np.argsort(xj.values)
    # Create a scatter plot of the predicted values
    
    #plt.plot(xj, pred)
    # Create a scatter plot of the actual values
    plt.scatter(xj, y.values, color="black", alpha=1/3)
    # Plot lower and upper quantile in dashed lines
    plt.plot(xj.values[order], qrs[order][:,0], color="C1", ls="--")
    plt.plot(xj.values[order], qrs[order][:,1], color="C1", ls="--")
    # Paint the region between the quantiles
    plt.fill_between(xj.values[order].ravel(),
    qrs[order][:,0].ravel(),
    qrs[order][:,1].ravel(),
    alpha=0.2,
    )
    # Label the x-axis with the feature name
    plt.xlabel(feature_name)
    plt.ylabel("predicted rent per sqm")
```

```python
import matplotlib.pyplot as plt 

plot_quantile_by_feature(X_new, y_new, y_pred, y_pis, "area")
```

```python
from mapie.metrics import regression_coverage_score


cov = regression_coverage_score(y_new, y_pis[:,0], y_pis[:,1])
print("Coverage: {:.2%}".format(cov))
```

#### Machine Learning Interpretability.

The field of ML Interpretability aims to dissect a ML black box, drilling down predictions and its learning process by powerfull agnostic tools. This methods are lumped in two big classes of methods: Global Model-Agnostic and Local Model-Agnostic.

-----
Global Model-Agnostic.   

1. PDP. Partial dependence plot shows the marginal effect one or two features have on the predicted outcome.    



```python
lgbm_model = LGBMRegressor()

lgbm_model.fit(X_train, y_train)
```

```python
from sklearn.inspection import PartialDependenceDisplay

PartialDependenceDisplay.from_estimator(lgbm_model, X_calib, X_calib.columns, kind='both')
```

```python
PartialDependenceDisplay.from_estimator(lgbm_model, X_calib, X_calib.columns, kind='both', centered=True)
```

2. Permutation Feature Importance. A great technique that measures the increase of prediction errors by means of permutation of feature values.

```python
from sklearn.inspection import permutation_importance


result = permutation_importance(
    lgbm_model, X_calib, y_calib, n_repeats=10, random_state=42, n_jobs=2
)

lgbm_permutation_feature_importance = pd.Series(result.importances_mean, index=X_calib.columns)


fig, ax = plt.subplots()
lgbm_permutation_feature_importance.plot.bar(yerr=result.importances_std)
ax.set_ylabel("Mean Error Increasing")
```

3. Model's feature importance 

```python
from lightgbm import plot_importance

plot_importance(lgbm_model, importance_type="gain")
```

2. Counterfactual Explanations

In interpretable machine learning, counterfactual explanations can be used to explain predictions of individual instances. The “event” is the predicted outcome of an instance, the “causes” are the particular feature values of this instance that were input to the model and “caused” a certain prediction.

https://christophm.github.io/interpretable-ml-book/counterfactual.html

```python
import dice_ml
from dice_ml import Dice

train_data = X_train.copy()
train_data["target"] = y_train

d_housing = dice_ml.Data(dataframe=train_data, 
                         continuous_features=X_train.columns.tolist(),
                         outcome_name="target")
# We provide the type of model as a parameter (model_type)
m_housing = dice_ml.Model(model=lgbm_model, backend="sklearn", model_type='regressor')
```

```python
exp_genetic_housing = Dice(d_housing, m_housing, method="genetic") 
```

```python
 X_calib[0:1]
```

```python
lgbm_model.predict( X_calib[0:1])
```

```python
# Multiple queries can be given as input at once
query_instances_housing = X_calib[0:1]
genetic_housing = exp_genetic_housing.generate_counterfactuals(query_instances_housing,
                                                               total_CFs=3,
                                                               features_to_vary=["area"],
                                                               desired_range=[10, 11])
genetic_housing.visualize_as_dataframe(show_only_changes=True)
```
